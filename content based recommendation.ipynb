## based on his own previous choices

```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df_users = pd.read_csv('users.csv')
df_posts = pd.read_csv('posts.csv')
df_views = pd.read_csv('views.csv')
```

```python
df_posts.columns=["post_id","title","category","post_type"]
df_posts=df_posts.dropna(how='any',axis=0) 
df_posts
```

```python
df_views
```

```python
df_users.columns=["user_id","name","gender","academics"]
df_users
```

```python
df_new=df_users.merge(df_views)
```

```python
df_new
```

```python
df_posts
```

```python
main_df=df_posts.merge(df_new)
```

```python
main_df
```

```python
main_df.describe()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def combine_features(row):
    return row['category'] +" "+row['post_type']
main_df["combined_features"] = main_df.apply(combine_features,axis=1)
```

```python
main_df
```

```python
cv = CountVectorizer()
```

```python
count_matrix = cv.fit_transform(main_df["combined_features"])
```

```python
count_matrix
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
cosine_sim
```

```python
_user="Nivesh Singh Chauhan"
```

```python
main_df['index'] = main_df.index
main_df
```

```python
def get_index_from_title(title):
    return main_df[main_df.name == title]["index"].values[1]
user_index = get_index_from_title(_user)
user_index
```

```python
similar_user=  list(enumerate(cosine_sim[user_index]))
similar_user
```

```python
sorted_similar_user = sorted(similar_user,key=lambda x:x[1],reverse=True)
```

```python
def get_title_from_index(index):
    return main_df[main_df.index == index]["title"].values[0]
f=[]
for element in sorted_similar_user:
    if element[1]<.5:
        continue
    f.append(get_title_from_index(element[0]))
for i in set(f):
    print(i)
```
